## Conversões

   * Conversão de número decimal para um número binário
   * Conversão de um número binário para um número decimal

```python
# Converte um valor decimal em um valor binário:
def converteBINARIO(num):
    print(f'\t--- Converter o numero {num} para binario  ---')
    ast()
    binario = []
    while( num > 0 ):
        sobra = int(num % 2)
        binario.insert(0, sobra)
        print(f'| {num} % 2 = {sobra}\t\t| {num} / 2 = {int(num/2)}')
        num = int(num / 2)
        
    print('\n')
    return binario
```

```python
# Converte um valor binário sem sinal em um valor decimal:
def converteDECIMAL(binario):
    # binario eh uma lista
    tam = len(binario)
    i = 0
    valor = 0
    
    while(tam > 0):
        pot = (2 ** i)
        mult = (binario[tam-1] * pot)
        valor = valor + mult
        print(f'| 2 elevado a {i} = {pot}\t| Valor = {valor}')
        tam = tam - 1
        i = i + 1
        
    return valor
```

## Números Binários e Representações com Sinais
   
   * Verificação de Overflow
   * Somar dois números binários
   * Sinal Magnitude
   * Complemento de 1
   * Complemento de 2

```python
def verificaOVERFLOW(binario1, binario2, binario3):
    # binario3 = resultado da soma
    # binario1 = operando 1
    # binario2 = operando 2
    
    # Regra do overflow: se dois números são somados e ambos são positivos ou ambos
    # negativos, então o overflow ocorre se, e somente se, o resultado tiver o 
    # sinal oposto.
    
    over = True
    if (binario1[0] == binario2[0]) and binario1[0] == 1:
        if binario3[0] == 0: 
            over = False
            print('Overflow causado!')
    elif (binario1[0] == binario2[0]) and binario1[0] == 0:
        if binario3[0] == 1: 
            over = False
            print('Overflow causado!')
            
    return over
```

```python
def soma(binario1, binario2):
    # realiza a soma de dois numeros binarios
    binario3 = []
    aux = 0
    n1 = int()
    n2 = int()
    for i in range(len(binario1)-1, -1, -1):
        n1 = binario1[i]
        n2 = binario2[i]
        if(n1 == 0 and n2 == 0 and aux == 0):
            aux = 0
            binario3.insert(0, 0)
        elif((n1 + n2 + aux) == 1):
            aux = 0
            binario3.insert(0, 1)
        elif((n1 + n2 + aux) == 2):
            aux = 1
            binario3.insert(0, 0)
        else:
            aux = 1
            binario3.insert(0, 1)
            
    over = verificaOVERFLOW(binario1, binario2, binario3)
   
    if over == False:
        resultado = over
    else:
        resultado = binario3

    return resultado
```

```python
# Retorna o número já na representação de sinal de magnitude
def sinal_magnitude(num):
    binario = converteBINARIO(num)
    while( len(binario) < 8 ):
        binario.insert(0, 0)
    return binario 
```

```python
# Inverte todos os bits do número binário
def complemento1(binario):
    binario2 = []
    for i in binario:
        if i == 1: 
            binario2.append(0)
        else: 
            binario2.append(1)
    return binario2
```

```python
def complemento2(num):
    # Realiza a conversão para a representação de complemento de 2
    num2 = abs(num)
    binario = sinal_magnitude(num2)
    binario = complemento1(binario)
    binario2 = sinal_magnitude(1)
    comp2 = soma(binario, binario2)
    return comp2
```

```python
def selecionaREP(num):
    # verifica qual a representação necessária para tal número.
    # Se o num seja positivo então é mandado para sinal magnitude
    # se for negativo manda o num para o complemento de 2
    if(num >= 0): 
        binario = sinal_magnitude(num)
    else: 
        binario = complemento2(num)
    return binario
```

## Funções Aritméticas

   * Inverter
   * Gestão de Inversão
   * Converter um número binário com sinal para decimal
   * Gestão de Somas
   * Subtração
   * Gestão de Subtração

```python
def inverter(num):
       
    # inverte um valor e retorna o número binário

    binario = complemento2(num)
    binario2 = sinal_magnitude(1)
    
    if num < 0:
        binario = complemento1(binario)
        binario = soma(binario, binario2)
    
    return binario
```

```python
def gestaoINV():
    # manda para a função de inversão e apresenta o resultado
    print('\t\t--- Função Inverter ---')
    num = lerNUM();
    binario = inverter(num)
    print('\n'*2)
    ast()
    print('Número invertido = ', end='')
    imprimeBIN(binario)
    print(f'\nQue eh equivalente a {num*(-1)}.\n')
```

```python
def converteDECIMAL_SINAL(binario):
    # recebe um número binario com sinal magnitude e converte para um número decimal
    
    if(binario[0] == 1):
        binario3 = sinal_magnitude(1)
        binario = complemento1(binario)
        binario = soma(binario, binario3)
        aux = binario[0]
        binario.pop(0)
        total = converteDECIMAL(binario)
        total = total * (-1)
        binario.insert(0, aux)
        
    else:
        aux = binario[0]
        binario.pop(0)
        total = converteDECIMAL(binario)
        binario.insert(0, aux)
    
    return total
```

```python
def gestaoSOMA():
    print('\t\t--- Somar 2 valores ---')
    ast()
    
    print()
    print('Numero 1: ')
    num1 = lerNUM()
    binario1 = selecionaREP(num1)
    
    print()
    print('Numero 2: ')
    num2 = lerNUM()
    binario2 = selecionaREP(num2)
    
    total = soma(binario1, binario2)
    
    if total != False:
        print('\n'*2)
        ast()
        print(f'total da soma = ', end='')
        imprimeBIN(total)
        print(f'\nQue eh equivalente a {num1 + num2}.\n')
```

```python
def subtracao(num1, num2):
    
    binario3 = sinal_magnitude(1)
    binario1 = selecionaREP(num1)
    
    if num2 > 0:
        binario2 = inverter(num2)
    else:
        binario2 = selecionaREP(num2)
    
    total = soma(binario1, binario2)
    
    return total
```

```python
def gestaoSUB():
    print('\t\t--- Subtracao ---')
    ast()
    
    print()
    print('Numero 1: ')
    num1 = lerNUM()
    
    print()
    print('Numero 2: ')
    num2 = lerNUM()
    
    binario = subtracao(num1, num2)
    
    if binario != False:
        print('\n'*2)
        ast()
        print('total da subtracao = ', end='')
        imprimeBIN(binario)
        print(f'\nQue eh equivalente a {num1-num2}.\n')
    
```
